```python
import pandas as pd
import numpy as np
pd.__version__
```

```python
df = pd.read_csv('Races.csv')
df[['Age','Gender']] = df['Age'].str.extract(r'(\d+)([ghm])')
df['Barrier'] = np.abs(df['Barrier'].astype(int))
df['Time'] = pd.to_timedelta('00:' + df['Time'])
```

```python
def format_deltas(timedelta):
    minutes, seconds = divmod(timedelta.seconds, 60)
    microseconds = timedelta.microseconds
    return f"{minutes:02d}:{seconds:02d}.{microseconds // 1000:04d}" 


means = df.groupby('Track')[['Time']].mean()
means['Time'] = means['Time'].apply(format_deltas)
means.sort_values('Time')
```

```python
df['Trainer'].value_counts().sort_values(ascending=False).to_frame().merge(df[['Trainer', 'From']], on='Trainer', how='inner').drop_duplicates().query('count>1')
```

```python
proportions = df.groupby(['From', 'Track'])[['Winner']].count().rename(columns={'Winner': '%'})
(100 * proportions / proportions.groupby(['From']).transform("sum")).round(1)

```
